```python
#integral alma
from scipy import integrate

x2=lambda x:x**2
I=integrate.quad(x2,0,6)
print(I)
```

```python
import scipy.integrate as integrate
import scipy.special as special

result=integrate.quad(lambda x: special.jv(2.5,x),0,4.5)
print(result)
```

```python
from scipy.misc import derivative

def f(x):
    return x**3 +x**2
D=derivative(f,1.0,dx=1e-6)

print(D)
```

```python
x=np.linspace(0,10,2000)
plt.figure(figsize=(10,6))
plt.plot(x,np.sin(x),label="$Sin(X)$")
plt.plot(x,np.cos(x),label="$Cos(X)$")
plt.xlabel(r'$X$',fontsize=18)
plt.ylabel(r'$Y$',fontsize=18)
plt.title("$Sin(X)$$ & $$ Cos(X)$",fontsize=14)
plt.legend(loc="upper right")
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(x,np.sin(x),label="$Sin(X)$")
plt.plot(x,np.cos(x),label="$Cos(X)$")
plt.fill_between(x,0,np.sin(x))
plt.xlabel(r'$X$',fontsize=18)
plt.ylabel(r'$Y$',fontsize=18)
plt.title("$Sin(X)$$ & $$ Cos(X)$",fontsize=14)
plt.legend(loc="upper right")
plt.show()
```

```python
plt.figure(figsize=(14,6))
x=np.linspace(0,10,100)
y1=np.sin(x)
y2=np.cos(x)
plt.subplot(1,2,1)
plt.plot(x,y1)
plt.subplot(1,2,2)
plt.plot(x,y2)
plt.show()
```

```python
plt.figure(figsize=(12,6))
x=np.linspace(0,10,100)
y1=np.sin(x)
y2=np.cos(x)
plt.subplot(2,1,1)
plt.plot(x,y1,"b")
plt.subplot(2,1,2)
plt.plot(x,y2,"r-")
plt.tight_layout()
plt.show()
```

```python
x=np.arange(-50,50)
y1=np.power(x,2)
y2=np.tan(x)
y3=np.exp(x)
y4=np.log10(x)

fig1,ax1=plt.subplots(nrows=2,ncols=2,figsize=(10,10))
ax1[0,0].plot(x,y1,"tab:blue")
ax1[0,0].set_title("square fnc")
ax1[0,0].set_xlabel(r'$X$',fontsize=18)
ax1[0,0].set_ylabel(r'$X$',fontsize=18)


ax1[0,1].plot(x,y2,"r")
ax1[0,1].set_title("tan fnc")
ax1[0,1].set_xlabel(r'$X$',fontsize=18)
ax1[0,1].set_ylabel(r'$X$',fontsize=18)


ax1[1,0].plot(x,y3,"g")
ax1[1,0].set_title("exp fnc")
ax1[1,0].set_xlabel(r'$X$',fontsize=18)
ax1[1,0].set_ylabel(r'$X$',fontsize=18)


ax1[1,1].plot(x,y4,"y")
ax1[1,1].set_title("log fnc")
ax1[1,1].set_xlabel(r'$X$',fontsize=18)
ax1[1,1].set_ylabel(r'$X$',fontsize=18)


plt.show()
```
