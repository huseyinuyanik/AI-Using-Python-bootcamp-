## scatter(sacılım)

```python
plt.style.use("seaborn-darkgrid")
```

```python
x1=np.array([250,159,350,252,450,550,455,358,158,355])
y1=np.array([40,50,80,90,100,50,60,88,54,45])
plt.scatter(x1,y1,color="r")
plt.xlabel("timespent",fontsize=14)
plt.ylabel("score",fontsize=14)
plt.title("scattergraph")
plt.show()
```

```python
x2=np.array([200,100,300,220,400,500,450,380,180,350])
y2=np.array([400,500,800,900,1000,500,600,808,504,405])

plt.scatter(x1,y1, label="Class1")
plt.scatter(x2,y2,label="Class2")
plt.xlabel("timespent",fontsize=14)
plt.ylabel("score",fontsize=14)
plt.title("scattergraph")
plt.legend(fontsize=12)
plt.show()
```

```python
plt.scatter(x1,y1, label="Class1",marker="o",color="m")
plt.scatter(x2,y2,label="Class2",marker="v")
plt.xlabel("timespent",fontsize=14)
plt.ylabel("score",fontsize=14)
plt.title("scattergraph")
plt.legend(fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(8,8))                   ####eksik kaldı
area=[48,30,20,15]
labels=["Low","Medium","High","Critical"]
total=np.sum(area)

def val_per(x):
    return {:.2f}%\n({:.0f}).format(x,total*x/100)

                      
colors=["#8BC34A","r","b","g"]
plt.pie(area, labels=labels,startangle=45, shadow="true", val_per="%1.1f%%" )
plt.show()
```

## Histogram

```python
x=np.random.normal(size=2000)
plt.hist(x,bins=50,color="yellowgreen")
plt.gca().set(title="hist", ylabel="frekans")
plt.show()
```

```python
x=np.random.rand(2000)
plt.hist(x,bins=50,color="orange", alpha=0.8)
plt.gca().set(title="hist", ylabel="frekans")
plt.show()
```

```python
plt.figure(figsize=(10,10))
x=np.random.normal(size=2000)
plt.hist(x,bins=40,color="yellowgreen",edgecolor="#6A9662")
plt.gca().set(title="hist", ylabel="frekans")
plt.show()
```

```python
plt.figure(figsize=(10,10))
x=np.random.normal(size=2000)
plt.hist(x,bins=30,color="yellowgreen",edgecolor="#6A9662")
plt.gca().set(title="hist", ylabel="frekans")
plt.show()



plt.figure(figsize=(10,10))
x=np.random.normal(size=2000)
plt.hist(x,bins=20,color="yellowgreen",edgecolor="#6A9662")
plt.gca().set(title="hist", ylabel="frekans")
plt.show()



plt.figure(figsize=(10,10))
x=np.random.normal(size=2000)
plt.hist(x,bins=10,color="yellowgreen",edgecolor="#6A9662")
plt.gca().set(title="hist", ylabel="frekans")
plt.show()
```

```python
plt.figure(figsize=(8,11))
x=np.random.normal(-4,1,size=800)
y=np.random.normal(0,1.5,size=800)
z=np.random.normal(3.5,1,size=800)


plt.hist(x,bins=30,color="yellowgreen",edgecolor="#6A9662",alpha=0.5)
plt.hist(y,bins=40,color="r",edgecolor="#6A9662", alpha=0.5)
plt.hist(z,bins=40,color="b",edgecolor="#6A9662", alpha=0.5)

plt.show()
```

```python
plt.figure(figsize=(10,10))
x=np.random.rand(2000)
plt.hist(x,bins=30,color="yellowgreen",edgecolor="#6A9662", cumulative=True)
plt.gca().set(title="hist", ylabel="frekans")
plt.show()
```
